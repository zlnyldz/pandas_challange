```python
# Calculate the total school budget and per capita spending per school from school_data
import pandas as pd

# Sample school_data DataFrame
data = {
    'School': ['School A', 'School B', 'School A', 'School B', 'School C'],
    'Total Students': [500, 300, 450, 350, 200],
    'Budget': [1000000, 750000, 1200000, 800000, 500000]
}

# Create the DataFrame
school_data = pd.DataFrame(data)

# Calculate the total budget per school
total_budget_per_school = school_data.groupby('School')['Budget'].sum().reset_index()

# Calculate the total student count per school
total_students_per_school = school_data.groupby('School')['Total Students'].sum().reset_index()

# Merge the total budget and total students DataFrames
school_summary = pd.merge(total_budget_per_school, total_students_per_school, on='School')

# Calculate per capita spending
school_summary['Per Capita Spending'] = school_summary['Budget'] / school_summary['Total Students']

# Rename columns for clarity
school_summary.columns = ['School', 'Total Budget', 'Total Students', 'Per Capita Spending']

# Display the result
(school_summary)
```

```python
# Use the provided code to calculate the passing rates
import pandas as pd

# Sample school_data_complete DataFrame
data = {
    'School': ['School A', 'School A', 'School B', 'School B', 'School C'],
    'Math Score': [85, 90, 78, 65, 88],
    'Reading Score': [88, 72, 75, 68, 85]
}

# Create the DataFrame
school_data_complete = pd.DataFrame(data)

# Calculate the total number of students per school
total_students = school_data_complete.groupby('School').size().reset_index(name='Total Students')

# Calculate the number of students passing math (score >= 70)
passing_math = school_data_complete[school_data_complete['Math Score'] >= 70]
passing_math_count = passing_math.groupby('School').size().reset_index(name='Passing Math')

# Calculate the number of students passing reading (score >= 70)
passing_reading = school_data_complete[school_data_complete['Reading Score'] >= 70]
passing_reading_count = passing_reading.groupby('School').size().reset_index(name='Passing Reading')

# Merge the counts with total students
passing_rates = total_students.merge(passing_math_count, on='School', how='left').fillna(0)
passing_rates = passing_rates.merge(passing_reading_count, on='School', how='left').fillna(0)

# Calculate passing rates
passing_rates['Passing Math Rate'] = (passing_rates['Passing Math'] / passing_rates['Total Students']) * 100
passing_rates['Passing Reading Rate'] = (passing_rates['Passing Reading'] / passing_rates['Total Students']) * 100

# Display the result
(passing_rates[['School', 'Total Students', 'Passing Math', 'Passing Math Rate', 'Passing Reading', 'Passing Reading Rate']])
```

```python
# Create a DataFrame called `per_school_summary` with columns for the calculations above.
import pandas as pd

# Sample school_data_complete DataFrame
data = {
    'School': ['School A', 'School A', 'School B', 'School B', 'School C'],
    'Math Score': [85, 90, 78, 65, 88],
    'Reading Score': [88, 72, 75, 68, 85]
}

# Create the DataFrame
school_data_complete = pd.DataFrame(data)

# Calculate the total number of students per school
total_students = school_data_complete.groupby('School').size().reset_index(name='Total Students')

# Calculate the number of students passing math (score >= 70)
passing_math = school_data_complete[school_data_complete['Math Score'] >= 70]
passing_math_count = passing_math.groupby('School').size().reset_index(name='Passing Math')

# Calculate the number of students passing reading (score >= 70)
passing_reading = school_data_complete[school_data_complete['Reading Score'] >= 70]
passing_reading_count = passing_reading.groupby('School').size().reset_index(name='Passing Reading')

# Merge the counts with total students
per_school_summary = total_students.merge(passing_math_count, on='School', how='left').fillna(0)
per_school_summary = per_school_summary.merge(passing_reading_count, on='School', how='left').fillna(0)

# Calculate passing rates
per_school_summary['Passing Math Rate'] = (per_school_summary['Passing Math'] / per_school_summary['Total Students']) * 100
per_school_summary['Passing Reading Rate'] = (per_school_summary['Passing Reading'] / per_school_summary['Total Students']) * 100

# Rename columns for clarity
per_school_summary.columns = ['School', 'Total Students', 'Passing Math', 'Passing Reading', 'Passing Math Rate', 'Passing Reading Rate']

# Display the per_school_summary DataFrame
(per_school_summary)
```

```python
# Create a DataFrame called `size_summary` that breaks down school performance based on school size (small, medium, or large).
# Use the scores above to create a new DataFrame called `size_summary`
import pandas as pd

# Example DataFrame with school data
per_school_summary = pd.DataFrame({
    "School Name": ["School A", "School B", "School C", "School D"],
    "Total Students": [500, 1500, 2500, 4000],
    "Average Math Score": [85, 90, 78, 88],
    "Average Reading Score": [88, 92, 80, 85],
})

# Define the bins and labels for school sizes
size_bins = [0, 1000, 2000, 5000]  # Define the bin edges
size_labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]  # Define the corresponding labels

# Categorize school sizes
per_school_summary['School Size'] = pd.cut(per_school_summary['Total Students'], bins=size_bins, labels=size_labels, right=False)

# Create the `size_summary` DataFrame by grouping by 'School Size' and calculating the averages
size_summary = per_school_summary.groupby('School Size').agg({
    'Average Math Score': 'mean',
    'Average Reading Score': 'mean',
    'Total Students': 'mean'
}).reset_index()

# Display the `size_summary` DataFrame
print(size_summary)

# Display results
size_summary
```

```python
# Group the per_school_summary DataFrame by "School Type" and average the results.
import pandas as pd

# Example DataFrame with school data, including School Type
per_school_summary = pd.DataFrame({
    "School Name": ["School A", "School B", "School C", "School D"],
    "School Type": ["Public", "Private", "Public", "Private"],
    "Total Students": [500, 1500, 2500, 4000],
    "Average Math Score": [85, 90, 78, 88],
    "Average Reading Score": [88, 92, 80, 85],
})

# Group the DataFrame by "School Type" and calculate the averages
type_summary = per_school_summary.groupby("School Type").mean(numeric_only=True).reset_index()

# Display the resulting DataFrame
print(type_summary)
```
